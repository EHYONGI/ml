```python
import pandas as pd
import numpy as np
train_data = pd.read_csv("data/train.csv")
test_data = pd.read_csv("data/test.csv")

train_input = train_data[['Pclass', 'Sex', 'SibSp', 'Parch]]
train_input['Sex'] = train_input['Sex'].map({'male':1, 'female':0})
train_target = train_data[['Survived']]

test_input = test_data[['Pclass', 'Sex', 'SibSp', 'Parch']]
test_input['Sex'] = test_input['Sex'].map({'male':1, 'female':0})

from sklearn.ensemble import RandomForestClassifier

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=42)
dt.fit(train_input, train_target)

from sklearn.model_selection import cross_validate
scores = cross_validate(dt, train_input, train_target)

from sklearn.model_selection import StratifiedKFold
splitter = StratifiedKFold(n_splits=10, shuffle=True)
scores = cross_validate(dt, train_input, train_target, cv=splitter)
np.mean(scores['test_score'])
scores
```

```python
from sklearn.model_selection import GridSearchCV
params = {
    'min_impurity_decrease' : np.arange(0.0001, 0.001, 0.0001),
    'max_depth' : range(5, 15, 1),
    'min_samples_split' : range(2, 100, 10)
}
dt =DecisionTreeClassifier(max_depth=5, min_impurity_decrease=0.0001, min_samples_split=2)

gs = GridSearchCV(dt, params, n_jobs=-1)
gs.fit(train_input, train_target)
```

```python
predictions = gs.predict(test_input)
predictions
```

```python
DecisionTreeClassifier(max_depth=5, min_impurity_decrease=0.0001, min_samples_split=2)
gs.best_score_

result = {
    'PassengerId': test_data['PassengerId'],
    'Survived': predictions,
}
output = pd.DataFrame(result)
output.head()
output.to_csv('my_submission.csv', index=False)
```
